### 1. 네이버 웹툰 이미지 다운로드
* Referer라는 header값에 url 설정
* 특정 image url을 list에 저장
* requests의 get() 함수로 image data 요청
* response.content라는 property사용
* local file로 저장

```python
import requests
import os # os라는 내부 모듈 import했지? 

# {}는 dict 선언이라구
req_header = {
    'referer':'https://comic.naver.com/webtoon/detail.nhn?titleId=626906&no=602&amp;weekday=tue'
} # referer는 정해진 header값이야
img_urls = [
    'https://image-comic.pstatic.net/webtoon/626906/602/20200720170605_25afe583d26a90598d9dc108ff5f5c98_IMAG01_1.jpg',
    'https://image-comic.pstatic.net/webtoon/626906/602/20200720170605_25afe583d26a90598d9dc108ff5f5c98_IMAG01_2.jpg',
    'https://image-comic.pstatic.net/webtoon/626906/602/20200720170605_25afe583d26a90598d9dc108ff5f5c98_IMAG01_3.jpg',
]
for img_url in img_urls:
    res = requests.get(img_url, headers=req_header) # 얘한테 headers라는 속성이 있어서 그걸 지정해줘
    # print(res.ok) 하면 정상적으로 가져왔는지 보는 것 같은데? True로 뜸
    # print(res.content) 하면 content찍는거긴 한데 binary data라서 뭐가 와르르륵 떠
    img_data = res.content # 이렇게 와르륵 뜬 binary data를 image로 저장할거라서 content라는 property 끌고옴
    file_name = os.path.basename(img_url)
    # print(file_name) 파일명만 쫙 가져왔지
    with open(file_name, 'wb') as file: # binary data라서 wb라는데?
        print('Writing to {} ({} byte)'.format(file_name, len(img_data))) # 앞에는 파일명, 뒤에는 사이즈를 주도록
        file.write(img_data)
```

### 2. 네이버 웹툰 이미지 업로드
* http://httpbin.org/post 사이트에 다운받은 이미지 세개 올릴것
* requests의 post() 함수를 사용하고, files 속성에 image data지정

```python
import requests

upload_files_dict = {
    
    'img1': open('familysize_01.jpg', 'rb'),
    'img2': open('familysize_02.jpg', 'rb'),
    'img3': open('familysize_03.jpg', 'rb')
}
url = 'http://httpbin.org/post'
res = requests.post(url, files = upload_files_dict)
res.status_code
```

```python
res.json() # 이거 먼저 쳐서 key값 확인하면... 앞에 key값들 보이지? 그중에 image를 보고싶은거니까 밑에서 호출해줄거야
```

```python
img1 = res.json()['files']['img1']
len(img1)
```
